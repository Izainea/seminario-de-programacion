## Algoritmo TSNE

El algorithmo TSNE (t-Distributed Stochastic Neighbor Embedding) es un algoritmo de reducción de dimensionalidad no lineal que se utiliza para visualizar datos de alta dimensionalidad en un espacio de menor dimensión.

La idea principal detrás de TSNE es que los puntos que están cerca en el espacio de alta dimensión deben estar cerca en el espacio de baja dimensión y los puntos que están lejos en el espacio de alta dimensión deben estar lejos en el espacio de baja dimensión.

El algoritmo TSNE tiene dos etapas:

1. En la primera etapa, TSNE calcula la probabilidad de similitud entre pares de puntos en el espacio de alta dimensión. La similitud entre dos puntos se calcula como la probabilidad condicional de que los dos puntos sean vecinos más cercanos si se muestrean de acuerdo con una distribución de probabilidad gaussiana centrada en el punto.

2. En la segunda etapa, TSNE mapea los puntos de alta dimensión a un espacio de baja dimensión de manera que la similitud entre los puntos en el espacio de baja dimensión sea lo más cercana posible a la similitud entre los puntos en el espacio de alta dimensión.



# Aplicación de t-SNE al conjunto de datos de dígitos manuscritos
Este cuaderno muestra cómo aplicar el método t-SNE para la reducción de dimensionalidad en el conjunto de datos de dígitos manuscritos y cómo visualizar los resultados.


```python
# Importamos las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.manifold import TSNE
```

## Cargar el Conjunto de Datos

```python
# Cargamos el conjunto de datos de dígitos manuscritos
digits = datasets.load_digits()
```

## Exploración de los Datos

```python
# Datos de las imágenes y las etiquetas
X = digits.data  # Datos de las imágenes, cada una es un vector de 64 elementos
y = digits.target  # Etiquetas de los dígitos, el número que representan

# Mostramos las dimensiones de los datos
print(f"Dimensiones de los datos de las imágenes: {X.shape}")
print(f"Dimensiones de las etiquetas: {y.shape}")
```

```python
X[300].reshape(8, 8)

plt.imshow(X[589].reshape(8, 8), cmap='gray')
y[589]
```

```python
## Visualización de los datos

# Creamos una figura para mostrar las imágenes

fig, axs = plt.subplots(2, 5, figsize=(10, 5))

for i in range(10):
    ax = axs[i // 5, i % 5]
    ax.imshow(X[i].reshape(8, 8), cmap="gray")
    ax.axis('off')
    ax.set_title(f"Etiqueta: {y[i]}")
```

## Aplicación de t-SNE

```python
# Configuración de t-SNE con 2 componentes
tsne_1 = TSNE(n_components=2, random_state=2340)
tsne_2 = TSNE(n_components=2, random_state=2340)

# Reducción de la dimensionalidad
X_2d_1 = tsne_1.fit_transform(X_1)

X_2d_1
```

## Visualización de los Resultados

```python
# Configuramos el tamaño de la figura
plt.figure(figsize=(10, 8))

# Colores para los dígitos del 0 al 9
colors = ['red', 'green', 'blue', 'cyan', 'magenta', 'yellow', 'black', 'orange', 'darkgrey', 'lime']

# Dibujamos los puntos
for i, color in enumerate(colors):
    plt.scatter(X_2d_1[y_1 == i, 0], X_2d_1[y_1 == i, 1], c=color, label=i, alpha=0.5)
   # plt.scatter(X_2d_2[y_2 == i, 0], X_2d_2[y_2 == i, 1], c=color, label=i, alpha=0.5)

# Añadimos la leyenda y los títulos
plt.legend()
plt.title('t-SNE de dígitos manuscritos')
plt.xlabel('Componente t-SNE 1')
plt.ylabel('Componente t-SNE 2')

# Mostramos el gráfico
plt.show()
```
